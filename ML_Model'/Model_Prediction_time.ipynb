```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
models = ['Decision Tree', 'Random Forest', 'XGBoost', 'Phreeqpy']

prediction_times_1 = [0.0002026808261871338, 0.00040007519721984864, 0.00018621087074279785, 0.05]
prediction_times_1000 = [0.00036853885650634763, 0.0008913354873657227, 0.0003401286602020264, 8.65]
prediction_times_10000 = [0.000485623836517334, 0.0010883681774139405, 0.00048383164405822754, 90.56]

# Updating the x-axis labels to include PHREEQpy times
simulations = ['(1)', '(1000)', '(10000)']
x = np.arange(len(simulations))  # x-axis positions for the groups

# Times for each model
dt_times = [prediction_times_1[0], prediction_times_1000[0], prediction_times_10000[0]]
rf_times = [prediction_times_1[1], prediction_times_1000[1], prediction_times_10000[1]]
xgb_times = [prediction_times_1[2], prediction_times_1000[2], prediction_times_10000[2]]
phreeqpy_times = [prediction_times_1[3], prediction_times_1000[3], prediction_times_10000[3]]

# Plotting
fig, ax = plt.subplots(figsize=(14, 6))

width = 0.2  # width of the bars

# Bar plots
rects1 = ax.bar(x - 1.5 * width, dt_times, width, label='Decision Tree')
rects2 = ax.bar(x - 0.5 * width, rf_times, width, label='Random Forest')
rects3 = ax.bar(x + 0.5 * width, xgb_times, width, label='XGBoost')
rects4 = ax.bar(x + 1.5 * width, phreeqpy_times, width, label='Phreeqpy')

# Adding labels and title
ax.set_xlabel('Simulation')
ax.set_ylabel('Time (seconds)')
ax.set_title('Compare ML models time vs Phreeqpy')
ax.set_xticks(x)
ax.set_xticklabels(simulations)
ax.set_yscale('log')  # Setting the y-axis to log scale
ax.legend()

# Display the plot
plt.show()
```
